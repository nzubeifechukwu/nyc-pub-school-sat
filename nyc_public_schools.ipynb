# Analyzing SAT Scores Across Public Schools in New York

In this project, I'll analyze SAT scores for public schools in New York City to identify the top-performing schools in math and overall. I'll also explore how performance varies by borough.

The idea and dataset for this are from [this DataCamp project](https://app.datacamp.com/learn/projects/exploring_nyc_public_school_test_result_scores/guided/Python).

## Load, inspect and prepare data

```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load and inspect data
schools = pd.read_csv('schools.csv')
print(schools.info())
schools.head()
```

Let's drop the **building_code** and **percent_tested** columns.

```python
schools.drop(columns=['building_code', 'percent_tested'], inplace=True)
schools.head()
```

## Distribution of schools across borough

```python
print(schools['borough'].value_counts())
plt.bar(x=schools['borough'].value_counts().index, height=schools['borough'].value_counts());
```

## Which schools have the best math results?

Let's define the best math scores as at least 80% of the maximum possible score, which is 800.

```python
best_math_schools = schools[schools['average_math'] >= 0.8 * 800].sort_values(
    'average_math', ascending=False)[['school_name', 'borough', 'average_math']]
best_math_schools
```

## What are the top 10 performing schools based on the combined SAT scores?

```python
schools['total_SAT'] = schools.sum(axis=1)
top_10_schools = schools.sort_values('total_SAT', ascending=False)[['school_name', 'borough', 'total_SAT']].head(10)
top_10_schools
```

A lot of the best-performing schools in math also do very well overall. Forty percent (40%) of the best schools in math and overall are from Manhattan. Staten Island borough, with only 10 schools of the total 375 (less than 3%), has a school in the best performing cohorts in math and across all subjects, which is impressive.

## How does performance vary by borough?

```python
summary_stats_by_borough = (
    schools
    .groupby('borough')['total_SAT']
    .describe()
    .sort_values('std', ascending=False)
    .round(2)
)
summary_stats_by_borough.columns = [
    'num_schools', 'average_SAT', 'std_SAT', 'min_SAT', '25%_SAT', '50%_SAT', '75%_SAT', 'max_SAT'
]
summary_stats_by_borough.sort_values('std_SAT', ascending=False, inplace=True)
summary_stats_by_borough
```

```python
summary_stats_by_borough['median - mean'] = (
    summary_stats_by_borough['average_SAT'] - summary_stats_by_borough['50%_SAT']
).abs()
summary_stats_by_borough
```

The standard deviations from the mean SAT scores are not particularly large (the differences between the means and medians are also not large), suggesting that the average score for each borough is a reasonable indication of the general performance of schools in the borough.
